```python
# Import needed libraries
%matplotlib inline 
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import pickle
from IPython.display import display

DATA_DIR = "./"
EARTHQUAKE_PATH = './EARTHQUAKE.dat'
```

# Part I : Data

## Section A : Series

### Point a) Retrieve the series contained in the matrix.

```python
# Read the csv file and give the proper column names
# `hour` column represents the fraction of the day incident happened

df = pd.read_csv(EARTHQUAKE_PATH, delim_whitespace=True, names=['year', 'month', 'day', 'hour', 'magnitude'])
display(df)
```

### Point b) Get the date

```python
# gets the fraction of day and translates it in hours, minutes and seconds
def fraction_to_time(day_frac):
    hour = day_frac*24
    minute = (hour - int(hour))*60
    second = (minute - int(minute))*60
    return int(hour), int(minute), int(second)

# returns a datetime type from the row of the dataframe df
def get_datetime(row):
    hour, minute, second = fraction_to_time(row['hour'])
    x = datetime.datetime(int(row['year']), int(row['month']), int(row['day']), hour, minute, second)
    return x

date = df.apply(get_datetime, axis=1)
display(date)
```

### Point c) Filtering the data with amplitude >= 6

```python
# function returns the matrix of values where the magnitude starts from @min_m between @start_year and @end_year
# df : dataframe from which the magnitude and year can be obtained
# matrix : matrix to which the filtering is applied
# min_m : minimum magnitude
# start_year : starting year
# end_year : ending year
# returns the filtered version of matrix
def filter_data(df, matrix, min_m, start_year, end_year):
    scope = (df['magnitude'] >= min_m) & (df['year'] >= start_year) & (df['year'] <= end_year)
    return matrix[scope]
filtered_df = filter_data(df, df, 6, 1910, 2017)
display(filtered_df)
```

### Point d) Create matrix B

```python
data = {'date': date, 'magnitude': df['magnitude']}
B = pd.DataFrame(data)
# filter B for magnitude >= 6 between 1910 and 2017. Reset indexes as well.
B = filter_data(df, B, 6, 1910, 2017).reset_index(drop=True)
B.head(3)
```

```python
# Save dataframe B for later use
import os
# Create DATA_DIR path if does not exist
if not os.path.exists(DATA_DIR):
    os.makedirs(DATA_DIR)
pickle.dump(B, open(DATA_DIR + 'B.p', 'wb'))
```

### Point e) Plot the time series

```python
#Plot the earthquake cases in relations to magnitude and date of occurence
fig, ax = plt.subplots(figsize=(14, 4))

ax.plot(B['date'], B['magnitude'], '-', color='b', linewidth=0.1)

ax.set_title('Earthquake data from 1910 to 2017 (magnitudes > 6)', fontsize=14)
ax.set_xlabel('Time - Year', fontsize=12)
ax.set_ylabel('Magnitude', fontsize=12)
```

## Section B : Categorization

### Point a) Divide data into categories

```python
# return new df where magnitudes are in [min_m; max_m)
# df : dataframe
# min_m : minimum magnitude
# max_m : maximum magnitude
def classify_by_magnitude(df, min_m, max_m):
    index = (df['magnitude'] >= min_m) & (df['magnitude'] < max_m)
    return df[index].reset_index(drop=True)

# Creates empty dataframe with columns as categories
def new_category_df():
    return pd.DataFrame(columns = ['cat1', 'cat2', 'cat3', 'cat4'])

# returns total number of earthquake incidents from dataframe df divided in 4 categories
def get_categories(df):
        return [classify_by_magnitude(df, 6, 7).shape[0], 
                classify_by_magnitude(df, 7, 8).shape[0], 
                classify_by_magnitude(df, 8, 9).shape[0], 
                classify_by_magnitude(df, 9, 20).shape[0]
               ]

table2 = new_category_df()

# get total number (all years) of incidents per category
table2.loc['total'] = get_categories(B)
display(table2)
```

### Point b) Divide into categories and 10 year range

```python
# return new dataframe with rows where year is in [min_year, max_year)
def classify_by_year(df, min_year=1900, max_year=2030):
    min_date = datetime.datetime(min_year, 1, 1, 0, 0, 0)
    max_date = datetime.datetime(max_year, 1, 1, 0, 0, 0)
    index = (df['date'] >= min_date) & (df['date'] < max_date)
    return df[index].reset_index(drop=True)

# return the dataframe with year range as rows and earthquake category as columns
def get_year_range_table(df, start_date, end_date, step):
    table = new_category_df()
    for i in range(start_date, end_date, step):
        start, end = i, i + step
        if(end < end_date):
            table.loc[str(start)] = get_categories(classify_by_year(df, start, end))
        else:
            table.loc[str(start)] = get_categories(classify_by_year(df, start, end + 1))
    return table

# returns array of labels representing range. Example ["1910-1920", "1920-1930"]
def get_range_labels(start_date, end_date, step):
    return [str(i) + "-" + str(i+step) if i+step < end_date else str(i) + "-" + str(end_date) for i in range(start_date, end_date, step)]

start_date, end_date, step = 1910, 2017, 10
decade_range = get_year_range_table(B, start_date, end_date, step)

# Change index values as ranges to make table more precise and understandable
decade_range = decade_range.set_axis(get_range_labels(start_date, end_date, step), axis="index")
table2 = pd.concat([table2, decade_range])
table2
```

### Point c) Plot evolution of cases over time

```python
def make_datetime(year):
    return datetime.datetime(int(year), 1, 1, 0, 0, 0)

year_range = get_year_range_table(B, 1910, 2017, 1)

# Define matrix C
C = year_range.drop(columns=['cat4'])
# Move date from index to column
C = C.reset_index().rename(columns={'index': 'date'})
# convert date as string to datetime
C['date'] = C['date'].apply(make_datetime)
```

### c) Count the number and share of events between std zones

```python
def classify_by_magnitude(df, min_m, max_m):
    index = (df['magnitude'] >= min_m) & (df['magnitude'] < max_m)
    return df[index].reset_index(drop=True)

def count_in_magnitude_range(df, min_m, max_m):
    return classify_by_magnitude(df, min_m, max_m).shape[0]


std_distribution_dict = {}


# for x < std1
l_bound = table3['mean']-table3['std1']
h_bound = table3['mean']+table3['std1']

n_std1 = count_in_magnitude_range(B, l_bound, h_bound)
n_std1_share = n_std1/B.shape[0]
std_distribution_dict['x<std1'] = [n_std1, n_std1_share]


# for std1 < x < std2
l_bound = table3['mean']-table3['std2']
h_bound = table3['mean']-table3['std1']
n_std2_below = count_in_magnitude_range(B, l_bound, h_bound)

l_bound = table3['mean']+table3['std1']
h_bound = table3['mean']+table3['std2']
n_std2_above = count_in_magnitude_range(B, l_bound, h_bound)

n_std2 = n_std2_below + n_std2_above
n_std2_share = n_std2/B.shape[0]
std_distribution_dict['std1<=x<=std2'] = [n_std2, n_std2_share]


# for x > std3
l_bound = 0
h_bound = table3['mean']-table3['std3']
n_std3_below = count_in_magnitude_range(B, l_bound, h_bound)

l_bound = table3['mean']+table3['std3']
h_bound = 20
n_std3_above = count_in_magnitude_range(B, l_bound, h_bound)

n_std3 = n_std3_above + n_std3_below
n_std3_share = n_std3/B.shape[0]
std_distribution_dict['x>std3'] = [n_std3, n_std3_share]


std_distribution_df = pd.DataFrame(std_distribution_dict, index=['number', 'share']).T
std_distribution_df['number'] = std_distribution_df['number'].astype(int)
display(pd.DataFrame({'value': table3}))
display(std_distribution_df)
```

## Section B: Median

### a) Calculate median, q1, q3, interquartile range.

```python
table4_dict = {}
table4_dict['q1'], table4_dict['median'], table4_dict['q3'] = B['magnitude'].quantile([0.25,0.5,0.75])
table4_dict['iqr'] = table4_dict['q3'] - table4_dict['q1']

table4_df = pd.DataFrame(table4_dict, index=['value']).T
table4_df
```

It would also be great to know the lower and upper extreme values in order to understand the borders of values. 
Yule-Kendall index to understand the skewness of the distribution.

```python
# So to get them

def get_ykindex(q1, q2, q3):
    return ((q3-2*q2+q1))/q3-q1

table4_df.loc['lower extreme'] = B['magnitude'].min()
table4_df.loc['upper extreme'] = B['magnitude'].max()
table4_df.loc['yk index'] = get_ykindex(table4_df.loc['q1'], table4_df.loc['median'], table4_df.loc['q3'])
table4_df
```

### b) Plot series, median, q1 and q3

```python
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(B['date'], B['magnitude'], 'grey', linewidth = 0.1)

ax.axhline(y=table4_df.loc['median'][0], linestyle = 'dashed', color='r', linewidth=2, label="median")
ax.axhline(y=table4_df.loc['q1'][0], linestyle = 'dashed', color='green', linewidth=2, label="q1")
ax.axhline(y=table4_df.loc['q3'][0], linestyle = 'dashed', color='orange', linewidth=2, label="q3")

ax.set_title('Earthquake data from 1910 to 2017 (magnitudes > 6)', fontsize=18)
ax.set_xlabel('Time - Year', fontsize=16)
ax.set_ylabel('Magnitude', fontsize=16)
ax.legend(fontsize=12)
```

### c) Count number of samples and their share between quantiles

```python
def get_share(n, total):
    return n/total
below_q1 = count_in_magnitude_range(B, 0, table4_df.loc['q1'][0])
above_q3 = count_in_magnitude_range(B, table4_df.loc['q3'][0], 20)
between_q1_q3 = count_in_magnitude_range(B, table4_df.loc['q1'][0], table4_df.loc['q3'][0])

quantile_dist_dict = {'below_q1': [below_q1, get_share(below_q1, B.shape[0])], 'between_q1_q3': [between_q1_q3, get_share(between_q1_q3, B.shape[0])], 'above_q3': [above_q3, get_share(above_q3, B.shape[0])]}

quantile_dist_df = pd.DataFrame(quantile_dist_dict, index=['value', 'share']).T

display(quantile_dist_df)
```

### d) Is this a normal distribution?
No, it is not. Although the percentage of samples between quantiles is very close to theoretical one, it is important to mention that the distribution is also positively (right) skewed. That can be noticed by the proximity of median to lower extreme as well as calculated Yule-Kendall index.

## Section C: Moving means and moving median

### a) Get moving mean, std1, std2, std3 with 10 year window and annual step

```python
moving_means={'date': [], 'mean': [], 'std1': [], 'std2': [], 'std3': []}
current_date = datetime.datetime(B['date'][0].year, 1, 1, 0, 0, 0)
last_date = B['date'][len(B)-1]
year_delta = datetime.timedelta(365)

while (current_date <= last_date):
    window = datetime.timedelta(365*10)
    index = (B['date'] >= current_date) & (B['date'] < current_date.replace(current_date.year + 10))
    mean = B[index]['magnitude'].mean()
    std = B[index]['magnitude'].std()
    moving_means['date'].append(current_date)
    moving_means['mean'].append(mean)
    moving_means['std1'].append(std)
    moving_means['std2'].append(std*2)
    moving_means['std3'].append(std*3)
    current_date = current_date.replace(current_date.year + 1)

    
moving_means_df = pd.DataFrame(moving_means)
moving_means_df
```

### b) Plot B series and moving mean, std1, std2, std3

```python
def get_q1(df) :
    return df['magnitude'].quantile(0.25)
def get_q3(df) :
    return df['magnitude'].quantile(0.75)
def get_iqr(df):
    return get_q3(df) - get_q1(df)

def get_bin_width(df, c=2.6):
    return c*get_iqr(df)/pow(len(df), 1/3)
```

## Section A: Adjustment example

```python
# get slope and intercept
def get_lin_reg_coef(x, y):
    x = np.array(x, dtype='int')
    y = np.array(y, dtype='float')
    model = np.polyfit(x, y, 1)
    a, b = model
    return a, b

# sort C for to fit model
C = classify_by_year(C, 1910, 2009)
# Convert date to ordinal for regression
date = pd.to_datetime(C['date'])
C['date_ordinal'] = date.map(datetime.datetime.toordinal)

cats = ['cat1', 'cat2', 'cat3']
cats_coef_dict = {}

for cat in cats:
    a, b = get_lin_reg_coef(C['date_ordinal'], C[cat])
    cats_coef_dict[cat] = [a, b]
cats_coef_df = pd.DataFrame(cats_coef_dict, index=['a', 'b'])
cats_coef_df
```

```python
# make dict for titles
cat_translate = {'cat1': 'category 1', 'cat2': 'category 2', 'cat3' : 'category 3'}
fig, axs = plt.subplots(ncols=3, figsize=(16, 3))
i = 0
# Plot the slopes
for cat in cats:
    a, b = cats_coef_df[cat].a, cats_coef_df[cat].b
    axs[i].plot(C['date'], C[cat])
    axs[i].plot(C['date_ordinal'], b + a*C['date_ordinal'], label='slope')
    axs[i].set_title(cat_translate[cat])
    axs[i].legend()
    i += 1
```
